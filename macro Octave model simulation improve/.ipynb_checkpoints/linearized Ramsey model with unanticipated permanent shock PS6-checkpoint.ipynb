### linearized solution of a Ramsey model with an unaticipated and  permanent shock to depreciation
### PS6 Q2
【recall 】CRRA utility function:

u(c) = (c^(1 -  η) - 1)/(1 -  η)

```python
import sympy as sym
import numpy as np
from sympy import lambdify
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
beta = 0.95
#discount factor
theta = 1.0/beta - 1.0
#discount rate
sigma = 2.0
#utility curvature
alpha = 0.3
#exponent on capital
Omega = 1
#productivity

delta_L = 0.1
#old depreciation rate
delta_H = 0.2
#new depreciation rate

T = 50
#number of time periods to simulate
```

```python
#initial steadystate
Kss_R1 = ( alpha * Omega / (theta + delta_L) )**(1/(1 - alpha))
Css_R1 = Omega * Kss_R1**alpha - delta_L * Kss_R1


#new steadystate
Kss_R2 = ( alpha * Omega / (theta + delta_H ))**(1/(1 - alpha))
Css_R2 = Omega * Kss_R2**alpha - delta_H * Kss_R2

```

```python
#utility function of ct
ct = sym.symbols('ct')
uct = (ct ** (1 - sigma) - 1 ) / (1 - sigma)
duct = sym.diff(uct)
dduct = sym.diff(duct)

#ct
uct = lambdify(ct, uct)
duct = lambdify(ct, duct)
dduct = lambdify(ct, dduct)

#production function of kt
kt = sym.symbols('kt')
fkt = kt ** alpha
dfkt = sym.diff(fkt)
ddfkt = sym.diff(dfkt)

#kt
fkt = lambdify(kt, fkt)
dfkt = lambdify(kt, dfkt)
ddfkt = lambdify(kt, ddfkt)
```

```python
#A matrix
coeff1_R2 = duct(Css_R2) * ddfkt(Kss_R2) / dduct(Css_R2)

A = np.zeros([2,2])
A[0, 0] = 1 + coeff1_R2/ (1 + theta)
A[0, 1] = -coeff1_R2
A[1, 0] = -1
A[1, 1] = 1 + theta
D_R2, V_R2 = LA.eig(A)
```

```python
#CHOOSE STABLE EIGENVALUE & VERIFY
A_time_V = np.around(A.dot(V_R2[:,D_R2<1]), decimals=8)
lambda_time_V = np.around( V_R2[:,D_R2<1]* D_R2[D_R2<1], decimals=8)
print(A_time_V == lambda_time_V)

determinant = (V_R2[0,D_R2<1] / V_R2[1,D_R2<1]).item()
lamda = D_R2.item(np.where(D_R2<1)[0][0])
```

```python
#Simulating dynamics 
deltak = Kss_R1 - Kss_R2
deltac = determinant * deltak 
kplot = []
cplot = []

for i in range(T):
    kplot.append(deltak)
    cplot.append(deltac)
    
    deltak = A[1, :].dot([deltac, deltak])
    deltac = determinant * deltak
    
c_dev = np.array(cplot)
k_dev = np.array(kplot)

#Computing levels
c_sim = np.array(cplot) + Css_R2
k_sim = np.array(kplot) + Kss_R2
```

```python
#Compute prices.
#Levels.
R_sim_level = [dfkt(i) for i in k_sim ]                                   
#Rental price of capital.
r_sim_level = [i - delta_H for i in  R_sim_level]                                     
#Real interest rate.
w_sim_level = [(fkt(k_sim[i]) - dfkt(k_sim[i]) * k_sim[i]) for i in range(len(k_sim))]
# or w_sim_level = [(fkt(i) - dfkt(i) * i) for i in k_sim]
#Wage rate.

rSS_R2 = theta
#derive from euler equation of household in steady state
RSS_R2 = rss_R2 + delta_H
wSS_R2 = fkt(Kss_R2) - dfkt(Kss_R2) * Kss_R2

#Deviations (from new steady state).
R_sim_dev = np.array(R_sim_level) - RSS_R2                  
r_sim_dev = np.array(r_sim_level) - rSS_R2
w_sim_dev = np.array(w_sim_level) - wSS_R2


#Compute Output
#Compute labor income share.
f = lambda x: fkt(x)
output = f(k_sim)
lab_inc_share_sim = w_sim_level/output

```
